# What is GitHub Copilot?

GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

# Basic Usage

## 1. Code Completion

Copilot provides suggestions as you type. Let's try some examples:

1. Start typing a function definition:
```python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
```

2. Copilot will suggest the implementation. Press Tab to accept or keep typing for different suggestions.

````python
def calculate_bmi(weight, height):
    # Calculate BMI using weight in kg and height in meters
    if height <= 0: 
        raise ValueError("Height must be greater than zero.")
    if weight <= 0:
        raise ValueError("Weight must be greater than zero.")
    bmi = weight / (height ** 2)
    return round(bmi, 2)
    # Classify BMI
def classify_bmi(bmi):
    # Classify BMI based on standard categories
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
def main():
    # Main function to get user input and display BMI classification
    try:
        weight = float(input("Enter weight in kg: "))
        height = float(input("Enter height in meters: "))
        bmi = calculate_bmi(weight, height)
        classification = classify_bmi(bmi)
        print(f"Your BMI is {bmi}. You are classified as: {classification}")
    except ValueError as e:
        print(e)
main()
    # Check if height and weight are valid
    
````

## 2. Comment-to-Code

One of Copilot's most powerful features is generating code from comments:


````python
# Create a function that converts temperature from Celsius to Fahrenheit
# Include input validation and round to 2 decimal places

# Try typing this comment and let Copilot suggest the implementation
def convert_celsius_to_fahrenheit(celsius):
    # Convert Celsius to Fahrenheit
    if celsius < -273.15:
        raise ValueError("Temperature cannot be below absolute zero.")
    fahrenheit = (celsius * 9/5) + 32
    return round(fahrenheit, 2)
    # Check if temperature is valid
def main_temp():    
    # Main function to get user input and display converted temperature
    try:
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = convert_celsius_to_fahrenheit(celsius)
        print(f"{celsius}°C is equal to {fahrenheit}°F")
    except ValueError as e:
        print(e)
main_temp()
````

## 3. Working with Libraries

Copilot is particularly helpful when working with libraries. It can suggest proper import statements and usage patterns:

````python
# Create a scatter plot using matplotlib with the following features:
# - Random data points
# - Custom colors based on point values
# - A trend line
# - Proper labels and title
# - A legend
# - Proper

# Let Copilot suggest the implementation
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider

# Generate random data points
np.random.seed(42)
x = np.random.rand(50) * 10  # Random x values
y = 2 * x + np.random.randn(50) * 5  # y values with some noise

# Initial slope and intercept
initial_slope = 2
initial_intercept = 0

# Create the figure and axis
fig, ax = plt.subplots()
plt.subplots_adjust(top=0.85, bottom=0.25)  # Adjust space for the slider

# Scatter plot
scatter = ax.scatter(x, y, c=y, cmap='viridis', label='Data Points')

# Plot the initial trend line
trend_line, = ax.plot(x, initial_slope * x + initial_intercept, color='red', label='Trend Line')

# Add color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Point Values')

# Add labels, title, and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scatter Plot with Adjustable Trend Line')
ax.legend()

# Add a slider for slope adjustment
ax_slider = plt.axes([0.2, 0.9, 0.65, 0.03])  # Position of the slider
slope_slider = Slider(ax_slider, 'Slope', -5.0, 5.0, valinit=initial_slope)

# Update function for the slider
def update(val):
    slope = slope_slider.val
    trend_line.set_ydata(slope * x + initial_intercept)  # Update trend line
    fig.canvas.draw_idle()

# Connect the slider to the update function
slope_slider.on_changed(update)

plt.show()
````


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    """
    A class to manage patient records with methods for adding, updating, retrieving,
    and performing basic statistical analysis on the records.
    """

    def __init__(self):
        self.records = {}

    def add_record(self, patient_id, name, age, condition):
        """
        Add a new patient record.

        Parameters:
        - patient_id (str): Unique identifier for the patient.
        - name (str): Name of the patient.
        - age (int): Age of the patient.
        - condition (str): Medical condition of the patient.

        Raises:
        - ValueError: If the patient_id already exists or invalid data is provided.
        """
        if patient_id in self.records:
            raise ValueError("Patient ID already exists.")
        if not isinstance(age, int) or age <= 0:
            raise ValueError("Age must be a positive integer.")
        self.records[patient_id] = {"name": name, "age": age, "condition": condition}

    def update_record(self, patient_id, **kwargs):
        """
        Update an existing patient record.

        Parameters:
        - patient_id (str): Unique identifier for the patient.
        - kwargs: Fields to update (e.g., name, age, condition).

        Raises:
        - KeyError: If the patient_id does not exist.
        """
        if patient_id not in self.records:
            raise KeyError("Patient ID not found.")
        self.records[patient_id].update(kwargs)

    def get_record(self, patient_id):
        """
        Retrieve a patient record.

        Parameters:
        - patient_id (str): Unique identifier for the patient.

        Returns:
        - dict: Patient record.

        Raises:
        - KeyError: If the patient_id does not exist.
        """
        if patient_id not in self.records:
            raise KeyError("Patient ID not found.")
        return self.records[patient_id]

    def get_average_age(self):
        """
        Calculate the average age of all patients.

        Returns:
        - float: Average age of patients.
        """
        if not self.records:
            return 0
        total_age = sum(record["age"] for record in self.records.values())
        return total_age / len(self.records)

    def get_condition_statistics(self):
        """
        Get statistics on the number of patients per condition.

        Returns:
        - dict: A dictionary with conditions as keys and counts as values.
        """
        stats = {}
        for record in self.records.values():
            condition = record["condition"]
            stats[condition] = stats.get(condition, 0) + 1
        return stats


# Example usage
records = PatientRecords()
records.add_record("P001", "Alice", 30, "Diabetes")
records.add_record("P002", "Bob", 45, "Hypertension")
records.update_record("P001", age=31)
print(records.get_record("P001"))
print("Average Age:", records.get_average_age())
print("Condition Statistics:", records.get_condition_statistics())
````

### ✏️ Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Sample data
data = {
    "Category": ["A", "B", "C", "D", "E"],
    "Values1": [10, 20, 15, 25, 30],
    "Values2": [5, 15, 10, 20, 25],
}
df = pd.DataFrame(data)

# Function to create plots
def create_dashboard(plot_type):
    plt.figure(figsize=(8, 6))
    
    if plot_type == "Bar Plot":
        df.plot(x="Category", y=["Values1", "Values2"], kind="bar", ax=plt.gca())
        plt.title("Bar Plot")
    elif plot_type == "Line Plot":
        df.plot(x="Category", y=["Values1", "Values2"], kind="line", ax=plt.gca())
        plt.title("Line Plot")
    elif plot_type == "Scatter Plot":
        plt.scatter(df["Values1"], df["Values2"], c="blue", label="Values1 vs Values2")
        plt.xlabel("Values1")
        plt.ylabel("Values2")
        plt.title("Scatter Plot")
        plt.legend()
    else:
        plt.text(0.5, 0.5, "Invalid Plot Type", fontsize=12, ha="center")
    
    plt.grid(True)
    plt.show()

# Interactive widget
interact(create_dashboard, plot_type=["Bar Plot", "Line Plot", "Scatter Plot"])
````
